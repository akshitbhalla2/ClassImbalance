```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from plot_metric.functions import BinaryClassification 
```

```python
# import dataset
data = pd.read_csv("creditcard.csv")
```

```python
# check the shape of dataset
data.shape
```

```python
# check head of the dataset
data.head()
```

```python
# check information about the dataset like - missing values, datatypes etc
data.isna().mean()*100
data.dtypes
```

```python
# check the target class distribution
data["Class"].value_counts()
```

```python
# create a visual plot to see the target distribution
sns.countplot(data["Class"])
```

```python
# 0 will be represented much more than 1 
# We might see 0 where we should expect 1
# i.e. We might observe False Negatives and must control them.
# Thus we must maximize RECALL (minimize False Negatives).
```

```python
# create some scatterplots based on features and see if you can see some pattern in the data
sns.pairplot(
    data, 
    diag_kind = "kde",
    hue = "Class"
)
```

```python
# baseline accuracy of the model
round(data['Class'].value_counts(normalize=True) * 100, 2)
```

```python
# drop the ID variable as it is unique for all the transactions and does not have any meaningful information about the data
data = data.drop(columns = ["ID"])
```

```python
# extract features and target from the original dataset
features = data.drop(columns = ["Class"])
target = data["Class"]
```

```python
# split the dataset into trainig and test set to train and evaluate the model respectively
X_train, X_test, Y_train, Y_test = train_test_split(
    features,
    target,
    test_size = 0.3,
    random_state = 42
)
```

```python
# create a random forest model
model = RandomForestClassifier(random_state = 42)
model = model.fit(X_train, Y_train)
```

```python
# predict given test dataset
pred = model.predict(X_test)
```

```python
# plot the confusion matrix
sns.heatmap(
    confusion_matrix(Y_test, pred), 
    annot = True,
    fmt = "0.0f"
)
```

```python
# Precision-recall curve
plot_precision_recall_curve(model, X_test, Y_test)
```

```python
# print the classification report
print(classification_report(Y_test, pred))
```

#### Synthetic Minority Oversampling Technique (SMOTE)

![alt text](images/smote.png "SMOTE")

```python
# import imblearn library and resample the original data using SMOTE technique
smote = SMOTE(random_state = 42)
X_smote, Y_smote = smote.fit_resample(X_train, Y_train)
```

```python
# train a random forest model on SMOTE data
model_smote = RandomForestClassifier(random_state = 42)
model_smote = model_smote.fit(X_smote, Y_smote) 
```

```python
# predict the classes on test data using model built on SMOTE data and plot the confusion matrix
pred_smote = model_smote.predict(X_test)
```

```python
# plot the confusion matrix
sns.heatmap(
    confusion_matrix(Y_test, pred_smote), 
    annot = True,
    fmt = "0.0f"
)
```

```python
# Precision-recall curve
plot_precision_recall_curve(model_smote, X_test, Y_test)
```

```python
# print the classification report
print(classification_report(Y_test, pred_smote))
```

```python
# ROC for the model we developed on SMOTE data
model_probabilities = model_smote.predict_proba(X_test) 
# 0th column has probabilities for 1 and 1st column has probabilities for 0
# We are interested in probabilities for 0 only
proba_0 = model_probabilities[:, 1]

bc = BinaryClassification(Y_test, proba_0, labels = [0, 1]) # default threshold = 0.5
bc.plot_roc_curve()
plt.show()
```

```python
# let's use another probability threshold so that we can get to the elbow position in the above curve
bc = BinaryClassification(Y_test, proba_0, labels = [0, 1], threshold = 0.06) # default threshold = 0.5
bc.plot_roc_curve()
plt.show()
```

```python
# compare these probabilities against the probability threshold of 6% rather than the default threshold of 50%
pred_smote_new = (proba_0 >= 0.06)
```

```python
# plot the confusion matrix
sns.heatmap(
    confusion_matrix(Y_test, pred_smote_new), 
    annot = True,
    fmt = "0.0f"
)
```

```python
# print the classification report
print(classification_report(Y_test, pred_smote_new))
```
